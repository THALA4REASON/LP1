```python
import os
```

```python
folder = "/content/drive/MyDrive/HTML"
```

```python
documents = {}
for file in os.listdir(folder):
    if file.endswith(".html"):
        with open(os.path.join(folder, file), "r", encoding="utf-8") as f:
            documents[file] = f.read().lower()

print("Indexed files:", list(documents.keys()))
```

```python
query = input("\nEnter your search query: ").lower()

results = []
for filename, content in documents.items():
    if query in content:
        results.append(filename)
```

```python
if results:
    print("\nüîç Search Results:")
    for r in results:
        print("‚Üí", r)
else:
    print("\nNo pages found matching your query.")
```
